```python
import pandas as pd
movies_df = pd.read_csv('Data/movies.csv')
ratings1_df = pd.read_csv('Data/ratings1.csv')
ratings2_df = pd.read_csv('Data/ratings2.csv')
dates_df = pd.read_csv('Data/dates.csv')
#display(movies_df.head(10))

#display(dates_df.info())
dates_df['date'] = pd.to_datetime(dates_df['date'], dayfirst= True)
#display(dates_df.head())
dates_df['date'].dt.year.mode()

#display(ratings1_df.info())
#display(ratings2_df.info())
ratings = pd.concat([ratings1_df, ratings2_df], ignore_index= True)
ratings = ratings.drop_duplicates(ignore_index=True)
#display(ratings)
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates_df.shape[0])
print('Число строк в таблице movies: ', movies_df.shape[0])
#print(ratings.shape[0] == dates_df.shape[0])

ratings_dates = pd.concat([ratings, dates_df], axis=1)
#display(ratings_dates.head(10))
#display(movies_df.head(10))

merged = ratings_dates.merge(
    movies_df,
    on='movieId',
    how='left'
)
display(merged.tail(1000))
merged.info()
```

```python
from itertools import count
from statistics import mean
import pandas as pd
ratings_movies_df = pd.read_csv('Data/ratings_movies.csv')

#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
    
#ratings_movies_df['date'] = pd.to_datetime(ratings_movies_df['date'])
ratings_movies_df['year_release'] = ratings_movies_df['title'].apply(get_year_release)

#ratings_movies_df[ratings_movies_df['year_release'] == 1999].groupby('title')['rating'].mean().sort_values() #Какой фильм, выпущенный в 1999 году,
# получил наименьшую среднюю оценку зрителей?
mask_gen = ratings_movies_df['year_release'] == 2010
#ratings_movies_df[mask_gen].groupby('genres')['rating'].mean().sort_values() #Какое сочетание жанров фильмов (genres), выпущенных в 2010 году,
#получило наименьшую среднюю оценку (rating)?

#ratings_movies_df.groupby('userId')['genres'].nunique().sort_values(ascending=False) #Какой пользователь (userId) посмотрел наибольшее количество 
#различных (уникальных) комбинаций жанров (genres) фильмов? 

#Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая.
#ratings_movies_df.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])


#Найдите сочетание жанров (genres) за 2018 году, которое имеет наибольший средний рейтинг (среднее по столбцу rating),
#и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.
mask_2018 = ratings_movies_df['year_release'] == 2018
gen_df = ratings_movies_df[mask_2018].groupby('genres')['rating'].agg(['mean', 'count'])
gen_df[gen_df['count'] > 10].sort_values(by='mean', ascending= False)


#Добавьте в таблицу новый признак year_rating — год выставления оценки. Создайте сводную таблицу, которая иллюстрирует 
#зависимость среднего рейтинга фильма от года выставления оценки и жанра. Выберите верные варианты ответа, исходя из построенной таблицы:
ratings_movies_df['date'] = pd.to_datetime(ratings_movies_df['date'])
ratings_movies_df['year_rating'] = ratings_movies_df['date'].dt.year

#display(ratings_movies_df.info())
#display(ratings_movies_df)

pivot_t = ratings_movies_df.pivot_table(
    values='rating',
    index='year_rating',
    columns='genres',
    fill_value=0,
    aggfunc='mean'
)
#pivot_t['Action|Adventure'].min() #За весь период (с 1996 по 2018 год) сочетание жанров Action|Adventure ни разу не получало среднюю оценку ниже 3. (да
#pivot_t['Action|Adventure|Animation|Children|Comedy|IMAX'].sort_values(ascending= False) #Наилучшую оценку жанр Action|Adventure|Animation|Children|Comedy|IMAX получил в 2010 году. (нет)
#pivot_t['Comedy'] #lля жанра Comedy прослеживается тенденция падения рейтинга с каждым годом (с 1996 по 2018).(нет)


#C Среди сочетаний жанров, получивших наивысшую среднюю оценку в 2018 году, есть сочетание Animation|Children|Mystery. (да)
pivot_b = ratings_movies_df.pivot_table(
    values='rating',
    index='genres',
    columns='year_rating',
    fill_value=0,
    aggfunc='mean'
)
pivot_b[2018].sort_values(ascending= False).head(20)
```
